```python
data = data[['NAME', 'Boro', 'ZipCode', '2017.0', '2018.0', '2019.0', '2020.0','2021.0']].groupby(by='ZipCode',as_index = False).sum()
#tripsByHour = data[['pickUpHour', 'DOLocationID']].groupby(by='pickUpHour', as_index=False).count()
```

```python
Years = ['2017.0','2018.0','2019.0','2020.0','2021.0']
```

```python
for x in range(0,4):
    data['{0}->{1}'.format(Years[x],Years[x+1])]=data[Years[x+1]]-data[Years[x]]
```

```python
for x in range(0,4):
    data['{0}->{1}PERCENT'.format(Years[x],Years[x+1])]=((data[Years[x+1]]-data[Years[x]])/data[Years[x]])*100
```

```python
data.columns
```

```python
Cols = {'2017.0', '2018.0', '2019.0', '2020.0',
       '2021.0', '2017.0->2018.0', '2018.0->2019.0', '2019.0->2020.0',
       '2020.0->2021.0', '2017.0->2018.0PERCENT', '2018.0->2019.0PERCENT',
       '2019.0->2020.0PERCENT', '2020.0->2021.0PERCENT'}
```

```python
CoolData = data
for col in Cols:
    CoolData['{0}_norm'.format(col)]= (CoolData['{0}'.format(col)] - CoolData['{0}'.format(col)].min()) / (CoolData['{0}'.format(col)].max() - CoolData['{0}'.format(col)].min())
data = CoolData    
```
